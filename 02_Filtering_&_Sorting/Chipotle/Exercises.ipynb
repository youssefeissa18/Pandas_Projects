### Step 1. Import the necessary libraries

```python
import numpy as np
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

```python
Url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
```

### Step 3. Assign it to a variable called chipo.

```python
chipo = pd.read_csv(Url,sep='\t')
chipo
```

```python
chipo.head()
```

### Step 4. How many products cost more than $10.00?

```python
chipo.dtypes
```

```python
chipo['item_price'] = chipo['item_price'].str.replace('$', '').astype(float) # make this line to convert price to float to make condition
chipo.dtypes['item_price']
```

```python
c = chipo[chipo['item_price'] > 10]
c
```

### Step 5. What is the price of each item?

```python
chipo[(chipo['item_name'] == 'Chicken Bowl') & (chipo['quantity'] == 1)]
```

### Step 6. Sort by the name of the item

```python
chipo.sort_values(by='item_name',ascending=False)
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
ch = chipo['quantity'].idxmax() # idmax use to return the max number depend the row i choice
most_expensive = chipo.loc[ch,'quantity']
most_expensive
```

```python
chipo[chipo['item_price'] == chipo['item_price'].max()]  # use this shape to appear the full row
```
