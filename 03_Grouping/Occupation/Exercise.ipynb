### Step 1. Import the necessary libraries

```python
import numpy as np
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
```

### Step 3. Assign it to a variable called users.

```python
users = pd.read_csv(url,sep='|')
users
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby('occupation')['age'].mean()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
male_count_per_occupation = users[users['gender'] == 'M'].groupby('occupation')['gender'].count() # count number of Male and calculate the ratio
total_count_per_occupation = users.groupby('occupation')['gender'].count()
male_ratio_per_occupation = (male_count_per_occupation / total_count_per_occupation)
male_ratio_per_occupation = male_ratio_per_occupation.sort_values(ascending=False)
male_ratio_per_occupation
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.groupby('occupation')['age'].agg(['max','min']) # multi groupby in same time
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(['occupation','gender'])['age'].mean()
```

### Step 8.  For each occupation present the percentage of women and men

```python
total_count_per_occupation_gender = users.groupby(['occupation', 'gender'])['gender'].count() # counter the sum of gender m and f
percentage_per_occupation_gender = (total_count_per_occupation_gender / total_count_per_occupation)*100
percentage_per_occupation_gender.unstack()
```

```python
total_count_per_occupation_gender = users.groupby(['occupation', 'gender'])['gender'].count()
percentage_per_occupation_gender = (total_count_per_occupation_gender / total_count_per_occupation)*100
percentage_per_occupation_gender
```
